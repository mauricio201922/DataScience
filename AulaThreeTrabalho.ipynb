##### Importando Numpy

```
import numpy as np
from matplotlib import pyplot as plt
```

##### Carregando o DataSet

```
data = np.genfromtxt('/content/drive/MyDrive/Ciência da Computação/4° Ano/1° Semestre/1° bimestre/DataScience/Aula 3/Trabalho/Dataset/iris.data.txt', delimiter=',', usecols=(0,1,2,3))
```

```
print(data[0:,0])
```

```
data[50:100]
```

##### Comparando iris Sertosa com iris Versicolor e montando um grafico de comparação do comprimento da Sépala

```
irisSertosa = data[0:50,0]
irisVersicolor = data[50:100,0]
xs = [i for i, _ in enumerate(irisSertosa)]
plt.plot(xs, irisSertosa, 'r.:', label='iris-sertosa')
plt.plot(xs, irisVersicolor, 'b.:', label='iris-versicolor')

plt.legend(loc=1)
plt.title('Comparando Iris do comprimento da Sepala')
plt.xlabel('Dados das linhas')
```

##### Comparando iris Sertosa com iris Virginica e montando um grafico de comparação do comprimento da Sépala

```
irisSertosa = data[0:50,0]
irisVirginica = data[100:150,0]
xs = [i for i, _ in enumerate(irisSertosa)]
plt.plot(xs, irisSertosa, 'y.:', label='iris-sertosa')
plt.plot(xs, irisVirginica, 'g.:', label='iris-virginica')

plt.legend(loc=1)
plt.title('Comparando Iris do comprimento da Sepala')
plt.xlabel('Dados das linhas')
```
